```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv(r'C:\Users\jesse.pisel\Downloads\Production 20180725 115522\Production 20180725 115522.csv',index_col=[9],
                  parse_dates=[9])
```

```python
data['Gas'].fillna(0.)
data['Gas'] = data['Gas'].str.replace(',', '')
data['Oil'].fillna(0.)
data['Oil'] = data['Oil'].str.replace(',', '')
```

```python
data = data[~data['Gas'].isnull()]
data = data[~data['Oil'].isnull()]
```

```python
data['Gas'] = data['Gas'].astype(int)
data['Oil'] = data['Oil'].astype(int)
```

```python
data = data.loc[:, data['Oil'].mean().sort_values(ascending=False).index]
```

```python
plt.rcParams['axes.facecolor'] = '#303030'
for i in range(51):
    data[data['Field'] == data.Field.unique()[i]].Oil.plot(alpha=0.4, figsize=(15,6))
```

```python
plt.rcParams['axes.facecolor'] = '#303030'
for i in range(51):
    data[data['Field'] == data.Field.unique()[i]].Gas.plot(alpha=0.4,  figsize=(15,6))
```
