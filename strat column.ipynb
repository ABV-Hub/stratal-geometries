```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib nbagg
```

```python
tops = pd.read_csv(r"D:\prbtops_inhouse.csv")
small = tops.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(tops.formation.values)
```

```python
tops["formation_numeric"] = le.transform(tops.formation.values)
```

```python
tops.api.unique()
```

```python
def stratColumn(api):
    bigColumn = []
    for well in api:
        column = []
        selected_well = tops[tops["api"] == well]
        column.insert(
            0,
            np.full(
                (np.ceil(selected_well.depth.values[0]).astype(int), 50), -1
            ),
        )
        for i in range(len(selected_well) - 1):
            thicknesses = np.ceil(np.diff(selected_well.depth)).astype(int)
            column.append(
                np.full(
                    (thicknesses[i], 50),
                    selected_well.formation_numeric.values[i],
                )
            )
        bigColumn.append(np.concatenate(column))
    plt.imshow(np.hstack(bigColumn))
```
